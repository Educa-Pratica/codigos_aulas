```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = int(50000) # Número de números aleatórios a serem gerados
semente = 42 # Semente para garantir a reprodutibilidade
np.random.seed(semente) # Define a semente para a geração de números aleatórios
raio = 1 # Raio unitário do circulo = raio do quadrado
```

```python
x = np.random.uniform(-raio, raio, N) # Gerar N números aleatórios entre -raio e raio para x
y = np.random.uniform(-raio, raio, N) # Gerar N números aleatórios entre -raio e raio para y
len(x), len(y)
```

```python
dentro = x**2 + y**2 <= raio**2 # Verificar quais pontos estão dentro do círculo
M = np.sum(dentro) # Contar quantos pontos estão dentro do círculo
M
```

```python
M/N * 100 # Probabilidade de, ao sortear um par x e y, esse par estar dentro do círculo
```

```python
pi_estimado = (M / N) * 4 # Estimar o valor de pi
print(f"Estimativa de pi: {pi_estimado}")
```

```python
fig, ax = plt.subplots();
ax.set_aspect('equal')
ax.add_patch(plt.Circle((0, 0), raio, fill=False, color='blue', label='Círculo'))

#plotar pontos com legenda dentro e fora
ax.scatter(x[dentro], y[dentro], color='green', label='Dentro do círculo', alpha=0.5)
ax.scatter(x[~dentro], y[~dentro], color='red', label='Fora do círculo', alpha=0.5)
ax.legend()
ax.set_xlim(-raio, raio)
ax.set_ylim(-raio, raio);
```
